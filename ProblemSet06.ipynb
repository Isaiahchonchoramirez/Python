The list _inp_ is a list of integers.  Write a while loop that does the following:

```python
inp = [71,111,130,32,66,108,117,127,101,33,256,71,111,1032,66,108,117,127,101]

# write your code below

message = ""
i = 0
while i < len(inp):
    num = inp[i]
    if num > 255:
        break
    elif num < 33 or num > 126:
        print(f"Skipped {num}")
    else:
        message += chr(num)
    i += 1
print(f"The message is {message}")
```

### <font color="blue">Problem 4</font>
Complete the function called gcf that calculates the Greatest Common Factor(Divisor) given two integers.  The function accepts two integers x and y as parameters and returns the GCF using the Euclidean algorithm below.  You need to use a while loop.

Step 1: r = x % y<br />
Step 2: x = y<br />
Step 3: y = r<br />
Step 4: Repeat Steps 1 through 3 while y is greater than 0<br />
Step 5: GCF = x<br />

Write code that iterates over the list number_pairs (you may use a for statement) and calls the function gcf with each pair and prints the output shown below.

```python
# write the function below

#Algorithm
def gcf(x, y):
    while y != 0:
        r = x % y
        x = y
        y = r
    return x
    
number_pairs = [(12,24),(24,48),(17,37)]

# write your code below

for pair in number_pairs:
    x, y = pair
    result = gcf(x, y)
    print(f"The greatest common factor of {pair} is {result}")
```

```python
dividend = [32,24,8,16]
divisor = [2,4,0,8]

# write your code below

for value in dividend:
    counter = 0
    while counter < len(divisor):
        try:
            result = value / divisor[counter]
            print(f'Dividing {value} by {divisor[counter]} has a result of {result}')
        except ZeroDivisionError:
            print(f'Dividing {value} by {divisor[counter]} is undefined')
        
        counter += 1
```

You are to write a password generator.  You code needs to do the following:

1. Prompt the user for a password length.
2. Use a while loop to iterate up to the password length.
3. In the loop generate a random character within the ASCII printable range.
4. Accumulate the character in the password.
5. After the loop terminates, print the password.

To generate a random character, you will need to import the _random_ module and use the _random.randrange_ function to generate a number between 48 and 122 (printable ASCII character range).  Convert the number to a character using the _chr_ function.  Accumulate that character in the password variable.

```python
import random

# initialize the pseudorandom number generator
random.seed()

print("Password Generator")

#  write your code below

password_length = int(input("How many characters would you like your password to be? "))

if password_length < 6:
    print("Please choose a password length of at least 6.")
else:
    password = []
    password.append(chr(random.randrange(48, 58)) )
    password.append(chr(random.randrange(65, 91)))
    password.append(chr(random.randrange(97, 123)))
    password.append(chr(random.choice([33, 35, 36, 37, 38, 40, 41, 42, 43, 45, 63, 64])))
    while len(password) < password_length:
        random_char = chr(random.randrange(32, 127))
        password.append(random_char)

    random.shuffle(password)

    final_password = ''.join(password)

    print(final_password)
```

### <font color="blue">Problem 10</font>
Write code that creates the comma separated values text file, called _Unit06_wnvhumancases.txt_, by iterating over the dictionary(_wnvDictByYear_) created in the previous problem.  Each line in the file must consist of three columns:  year, county, and number of cases for that county.  Below is an example showing some of the records.  Do not be concerned about the order of the records in the file.

        2018,Amador,1
        2018,Butte,6
        2018,Contra Costa,1
        ...
        2006,Tulare,8
        2006,Ventura,3
        2006,Yolo,26
        2006,Yuba,5


You need to use a _while_ loop to iterate over the dictionary.  The example below provides the necessary syntax:

         while wnvDictByYear:
           year, yearDict = wnvDictByYear.popitem()

The _popitem_ method returns the next key/value pair then removes that key/value pair from the dictionary.  This means that once the dictionary is emptied the loop terminates.  The implication is that if your code is rerun it would not produce any output because _wnvDictByYear_ is empty.  You must make your code repeatable without requiring rerunning the previous problem to rebuild _wnvDictByYear_. You may not copy the code from #9 to build _wnvDictByYear_. This is not as evident as you may think. You will need to do some research to find a solution.

```python
import csv
import copy
import os

def build_wnvDictByYear(file_path):
    wnvDictByYear = {}
    with open(file_path, 'r') as file:
        reader = csv.reader(file)
        next(reader)
        for row in reader:
            year = int(row[0])
            county = row[2]
            cases = int(row[3])
            if year not in wnvDictByYear:
                wnvDictByYear[year] = {}
            if county not in wnvDictByYear[year]:
                wnvDictByYear[year][county] = 0
            wnvDictByYear[year][county] += cases
    
    return wnvDictByYear

input_file_path = 'Unit04_wnvhumancases.csv'

wnvDictByYear = build_wnvDictByYear(input_file_path)

wnvDictCopy = copy.deepcopy(wnvDictByYear)

output_file_path = 'Unit06_wnvhumancases.txt'
with open(output_file_path, 'w', newline='') as file:
    writer = csv.writer(file)
    while wnvDictCopy:
        year, yearDict = wnvDictCopy.popitem()
        for county, cases in yearDict.items():
            writer.writerow([year, county, cases])

print(f"Data has been written to {output_file_path}")

print(f"\nContents of {output_file_path}:")
with open(output_file_path, 'r') as file:
    for line in file:
        print(line.strip())

           
```
