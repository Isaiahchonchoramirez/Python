```python
with open('population.json', 'r') as file:
    data = json.load(file)
world_population = [(int(record['year']), record['population']) for record in data if record['region'] == "WORLD"]

world_population.sort(key=lambda x: x[0])

print("Tuples sorted year and population for World.")
print(world_population)
```

```python
def printReport(sorted_population_list):
    print("Year-over-Year Population Growth Report:")
    print(f"{'Year':<10} {'Population':<15} {'Growth (%)':<10} {'Growth Chart':<10}")
    print("-" * 60)
    
    first_year, first_population = sorted_population_list[0]
    print(f"{first_year:<10} {first_population:<15,.2f} {'0.00':<10} {'':<10}")
    
    for i in range(1, len(sorted_population_list)):
        current_year, current_population = sorted_population_list[i]
        previous_year, previous_population = sorted_population_list[i - 1]
        
        growth_percentage = ((current_population - previous_population) / previous_population) * 100
        growth_chart = "*" * int(growth_percentage * 10)
        
        print(f"{current_year:<10} {current_population:<15,.2f} {growth_percentage:<10.2f} {growth_chart}")
printReport(world_population)
```

```python
def printReport(sorted_population_list):
    print("Year-over-Year Population Growth Report:")
    print(f"{'Year':<10} {'Population':<15} {'Growth (%)':<10} {'Growth Chart':<10}")
    print("-" * 60)
    
    first_year, first_population = sorted_population_list[0]
    print(f"{first_year:<10} {first_population:<15,.2f} {'0.00':<10} {'':<10}")
    
    for i in range(1, len(sorted_population_list)):
        current_year, current_population = sorted_population_list[i]
        previous_year, previous_population = sorted_population_list[i - 1]
        
        growth_percentage = ((current_population - previous_population) / previous_population) * 100
        growth_chart = "*" * int(growth_percentage * 10)
        
        print(f"{current_year:<10} {current_population:<15,.2f} {growth_percentage:<10.2f} {growth_chart}")
    
    first_population = sorted_population_list[0][1] / 1000000
    last_population = sorted_population_list[-1][1] / 1000000
    overall_growth_rate = ((last_population - first_population) / first_population) * 100
    
    print("The population growth rate between {} and {} is {:.3f}%.".format(
        sorted_population_list[0][0], sorted_population_list[-1][0], overall_growth_rate))
printReport(world_population)
```
